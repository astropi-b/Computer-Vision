## Genrating a Random 5x5 Grey scale image

```python
import numpy as np
import matplotlib.pyplot as plt

image = np.random.randint(0, 256, (5, 5))
plt.imshow(image, cmap='gray')
plt.show()
```

## Doing Deravitve along x and y as mentioned in class using Backword Diffrence method

### Defining the Filters along x and y directions

```python
filter_x = np.array([[-1, 1]])
filter_y = filter_x.T
```

### 0. Taking x and y dervatives of the orginal image separatly

```python
from scipy.signal import convolve2d

derivative_x = convolve2d(image, filter_x, mode='same')

derivative_y = convolve2d(image, filter_y, mode='same')
```

```python
plt.imshow(derivative_x, cmap='gray')
plt.title('Derivative x-axis')
plt.show()
```

```python
plt.imshow(derivative_y, cmap='gray')
plt.title('Derivative y-axis')
plt.show()
```

### 1. Taking x dervative of an image and then taking y dervative for it.

```python
result1=convolve2d(derivative_x, filter_y, mode='same')
```

```python
plt.imshow(result1, cmap='gray')
plt.title('Result 1')
plt.show()
```

### 2. Taking y dervative of an image and then taking x dervative for it.

```python
result2=convolve2d(derivative_y, filter_x, mode='same')
```

```python
plt.imshow(result2, cmap='gray')
plt.title('Result 2')
plt.show()
```

### 3.Comparing both the Results

```python
result1
```

```python
result2
```

```python
result1==result2
```

## We see here clearly the both end results are  same.
